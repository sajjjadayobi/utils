## N-Queens Problem
We don't look for a path. We are looking for the best configuration
- **Goal**: Place N queens on a N-by-N chess board so that no two queens can attack each other.
- Thus a solution requires that no two queens share the same row, column or diagonal.
- We want to reduce state space (**search complexity**)

### Problem formulation: First formulation
- **States**: each column contains only one queen.
- **Actions**: move a queen to another row in its column.
- **State Space Size**: $N^N$ (brute force)


### Problem formulation: Second formulation (Permuatation)
- **States**: each column and each row contains only one queen.
- **Actions**: move a queen to another row in its column which is not occupied.
- **State Space Size**: $N!$ (brute force)

## Hill Climbing (Optimization)
- It's greedy and isn't complete
- From this conf go to the best neighbor (local)

```python
initial_state = NQueensStatePermutation.random_state(N=4)
print(initial_state)
initial_state.plot(show_conflicts=True)
# number of negibors: n*(n-1) / 2
```

```python
for neighbor in initial_state.neighbors():
    print(neighbor)
```

```python
# each step (neighbors): swap the row of two queens
# - we can only change the row
# - but all rows are acupiued 
# go to the neighbor with the least confilict
next_state = initial_state.best_neighbor()
print(next_state)
next_state.plot(show_conflicts=True)
```

```python
best_neighbor = next_state.best_neighbor()
print(best_neighbor)
best_neighbor.plot(show_conflicts=True)
```

General Case

```python
class HillClimbing:    
    def search(self, state):
        current = state
        while True:
            current.plot(show_conflicts=True)
            neighbor = current.best_neighbor()
            if neighbor >= current: 
                return current
            current = neighbor
```

```python
initial_state = NQueensStatePermutation.random_state(N=30)
ha = HillClimbing()
solution = ha.search(initial_state)
solution
```
